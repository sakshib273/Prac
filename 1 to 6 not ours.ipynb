```r
# Step 1: Importing required libraries and dataset
library(lmtest)
library(car)
library(MASS)

# Loading the built-in Longley dataset
data(longley)

# Step 2: Defining variables
y <- longley$Employed
x1 <- longley$GNP
x2 <- longley$Unemployed
x3 <- longley$Armed.Forces
x4 <- longley$Population

# Step 3: Fitting the Multivariate Regression Model
model <- lm(Employed ~ GNP + Unemployed + Armed.Forces + Population,
            data = longley)

summary(model)

# Step 4: Checking overall model significance using F-Test
anova(model)

# Step 5: Checking assumptions using Diagnostic Plots
par(mfrow = c(2,2))
plot(model)

# Step 6: Individual Assumption Tests

# 6.1 Normality of residuals (Shapiro-Wilk Test)
shapiro.test(residuals(model))

# 6.2 Homoscedasticity (Breusch-Pagan Test)
bptest(model)

# 6.3 Multicollinearity (VIF)
vif(model)

# 6.4 Autocorrelation (Durbin-Watson Test)
dwtest(model)

# Step 7: Detecting Outliers using Cook’s Distance
cooksd <- cooks.distance(model)
plot(cooksd, type = "h", main = "Cook’s Distance", ylab = "Distance")
abline(h = 4/length(cooksd), col = "red")

# Step 8: Applying Remedial Measures (Removing Population)
model_refined <- lm(Employed ~ GNP + Unemployed + Armed.Forces, data = longley)
summary(model_refined)

# Step 9: Diagnostic Plots for Refined Model
par(mfrow = c(2,2))
plot(model_refined)

# Step 10: Additional Diagnostic Tests for the Refined Model
shapiro.test(residuals(model_refined))
bptest(model_refined)
vif(model_refined)
dwtest(model_refined)

# Step 11: Cook’s Distance for Refined Model
cooksd <- cooks.distance(model_refined)
plot(cooksd, type = "h", main = "Cook’s Distance", ylab = "Distance")
abline(h = 4/length(cooksd), col = "red")
```

```r
#p3#Step 1: Load dataset and libraries
library(lmtest)
library(car)

#Load the built-in 'cars' dataset
data("cars")
head(cars)

#Step 2: Fit a simple linear regression model
model=lm(dist ~ speed, data = cars)
summary(model)

#Step 3: Diagnostic Plots
par(mfrow = c(2,2))
plot(model)

#Step 4: Breusch-Pagan Test for Heteroscedasticity
bptest(model)
#Step 5: Shapiro-Wilk Test for Normality of Residuals
shapiro.test(residuals(model))
#Step 6: Durbin-Watson Test for Autocorrelation
dwtest(model)
#Step 7: Check for Outliers/Influential Points
cooksd <- cooks.distance(model)
plot(cooksd, type="h", main="Cook’s Distance", ylab="Distance")
abline(h = 4/length(cooksd), col="red")
#Step 8: Applying Remedial Measures
#Apply Log Transformation on the dependent variable
model_log=lm(log(dist) ~ speed, data = cars)
summary(model_log)
#Diagnostic Plots after Transformation
par(mfrow = c(2,2))
plot(model_log)
#Re-test for heteroscedasticity
bptest(model_log)
```

```r
#p4
 #Step 1: Load dataset and required libraries
data(mtcars)
head(mtcars)
library(car)
#Step 2: Fit the initial multiple regression model
model_full = lm(mpg ~ wt + hp + disp + drat, data = mtcars)
summary(model_full)
# Interpretation Guide:
# Check which predictors are significant and observe their p-values.
# Large standard errors or unexpected signs may indicate multicollinearity.
#Step 3: Check correlation among independent variables
cor_matrix = cor(mtcars[, c("wt", "hp", "disp", "drat")])
print(cor_matrix)
#Step 4: Calculate Variance Inflation Factor (VIF)
vif_full = vif(model_full)
print(vif_full)
#Step 5: Refit model after dropping a statistically insignificant and highly
correlated variable (disp)
# disp is often highly correlated with wt and hp
model_refined = lm(mpg ~ wt + hp + drat, data = mtcars)
summary(model_refined)
#Step 6: Recalculate VIF values for the refined model
vif_refined = vif(model_refined)
print(vif_refined)
#This step is optional just to demonstrate effectiveness of remedial measures
#Step 7: Compare VIF values before and after
comparison = data.frame(
Variable = names(vif_refined),
VIF_Before = vif_full[names(vif_refined)],
VIF_After = vif_refined
)
print(comparison)
```

```r
#p5
#Step 1: Load required libraries
library(lmtest)
library(dplyr)
#Step 2: Load dataset
data("AirPassengers")
AP=data.frame( Passengers = as.numeric(AirPassengers), Month =
as.numeric(time(AirPassengers)) )
head(AP)
#Step 3: Fit a simple regression model
model=lm(Passengers ~ Month, data = AP)
summary(model)
#Step 4: Plot residuals to visually inspect autocorrelation
plot(residuals(model), type = "o", main = "Residuals vs Time", xlab = "Time
Index", ylab = "Residuals")
abline(h = 0, col = "red")
#Step 5: Detect autocorrelation
#5.1 Durbin-Watson Test
dw_test=dwtest(model)
print(dw_test)
#5.2 Breusch-Godfrey Test (higher-order autocorrelation)
bg_test=bgtest(model, order = 5) # test up to lag 5
print(bg_test)
#Step 6: Create lagged variables
AP=AP %>% mutate( Passengers_Lag1 = dplyr::lag(Passengers, 1),
Passengers_Lag12 = dplyr::lag(Passengers, 12) )
#Remove rows with NA due to lagging
AP_lagged=na.omit(AP)
#Step 7: Fit regression model with lagged variables
model_lagged=lm(Passengers ~ Month + Passengers_Lag1 +
Passengers_Lag12, data = AP_lagged)
summary(model_lagged)
#Step 8: Plot residuals for lagged model
plot(residuals(model_lagged), type = "o", main = "Residuals vs Time (with
Lagged Variables)", xlab = "Time Index", ylab = "Residuals")
abline(h = 0, col = "red")
#Step 9: Detect autocorrelation for lagged model
dw_test_lagged=dwtest(model_lagged)
print(dw_test_lagged)
bg_test_lagged=bgtest(model_lagged, order = 5)
print(bg_test_lagged)
```
